```python
# Obtén la ruta de la carpeta que contiene los archivos .au
ruta = "genres"

# Para cada archivo .au en la carpeta y subcarpetas, conviértelo a .mp3
for raiz, directorios, archivos in os.walk(ruta):
    for archivo in archivos:
        if archivo.endswith(".au"):
            # Carga el archivo .au con pydub
            ruta_archivo = os.path.join(raiz, archivo)
            cancion = AudioSegment.from_file(ruta_archivo, format="au")
            
            # Guarda el archivo .mp3
            ruta_archivo_mp3 = os.path.splitext(ruta_archivo)[0] + ".mp3"
            cancion.export(ruta_archivo_mp3, format="mp3")
            print(ruta_archivo_mp3)
```

```python
from pydub import AudioSegment
import os

# Definimos la duración de los fragmentos en milisegundos
duration = 5000

# Ruta de la carpeta que contiene los archivos mp3
ruta = "genres"

# Recorremos la carpeta y subcarpetas en busca de archivos mp3
for raiz, directorios, archivos in os.walk(ruta):
    for archivo in archivos:
        if archivo.endswith(".mp3"):
            # Cargamos el archivo mp3 con pydub
            ruta_archivo = os.path.join(raiz, archivo)
            cancion = AudioSegment.from_file(ruta_archivo, format="mp3")
            
            # Obtenemos la duración de la canción en milisegundos
            duracion_cancion = len(cancion)
            
            # Fraccionamos la canción en fragmentos de duración "duration"
            for i in range(0, duracion_cancion, duration):
                fragmento = cancion[i:i+duration]
                
                # Guardamos el fragmento como un nuevo archivo mp3
                nombre_fragmento = os.path.splitext(archivo)[0] + f"_{i//duration}.mp3"
                ruta_fragmento = os.path.join(raiz, nombre_fragmento)
                fragmento.export(ruta_fragmento, format="mp3")
                print(ruta_fragmento)
```
